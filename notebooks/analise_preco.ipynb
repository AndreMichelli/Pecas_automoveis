## Tratamento dos dados do csv

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
df = pd.read_csv("../data/pecas.csv")

# Visualizar as primeiras linhas
print(df.head())

# Estatísticas descritivas
print(df.describe())

# Distribuição de preços para uma peça específica
peca_especifica = 'nome_da_peca'  # altere conforme o necessário
df_peca = df[df['peca'].str.lower() == peca_especifica.lower()]

plt.figure(figsize=(10, 6))
sns.histplot(df_peca['preco'], kde=True)
plt.title(f'Distribuição de preços para a peça: {peca_especifica}')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()
```
